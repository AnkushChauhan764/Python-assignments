```

def line(m,X_n,c):
  return np.array(np.dot(x,m)+c).reshape(82,1)

def error(m,X_n,c,y_a):
  y_p = line(m,X_n,c)
  return np.mean((y_a - y_p)**2)

def derivative_m(m,X_n,c,y_a):
  y_p = line(m,X_n,c)
  C = np.array((y_p-y_a)).reshape(82,1)
  return 2*np.array(np.mean(np.multiply(C,X_n),axis=0))

def derivative_c(m,X_n,c,y_a):
  y_p = line(m,X_n,c)
  C = np.array((y_p-y_a)).reshape(82,1)
  return 2*np.array(np.mean(C))

derivative_m(m,X_n,c,y_a)
```
